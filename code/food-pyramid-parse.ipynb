```python
import json
import numpy as np
import pandas as pd
import pickle
import psycopg2 as pg
import pandas.io.sql as pd_sql
import sqlalchemy
import xml.etree.ElementTree as ET
```

```python
# Read in an XML file to parse for output to SQL

tree = ET.parse('fns-mypyramid-food-raw-data/Food_Display_Table.xml')
root = tree.getroot()
```

```python
for child in root:
    print(child.tag, child.attrib)
```

```python
root[0][1].text
```

```python
len(root)
```

```python
for i in range(0, len(root)):
    for j in range(0, len(root[i])):
        print(root[i][j].tag, root[i][j].text)
```

```python
cols = []
for i in range(0, len(root[0])):
    cols.append(root[0][i].tag.lower())
```

```python
cols
```

```python
# Parse the data into a pandas dataframe

data = []
for i in range(0, len(root)):
    item = []
    for j in range(0, len(root[i])):
        item.append(root[i][j].text.lower())
    data.append(item)
nutrition = pd.DataFrame(data, columns=cols)
```

```python
# Output to SQL

engine = sqlalchemy.create_engine('postgresql://grant@localhost:5432/pyramid')
nutrition.to_sql('nutrition', con=engine, if_exists='append')
```
